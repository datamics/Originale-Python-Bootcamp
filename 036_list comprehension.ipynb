# Verständnis 

In Ergänzung zu Sequenz Anwendungen und Listen Methoden enthält Python eine etwas fortgeschrittenere Operation names list comprehension (Listen Verständnis).

list comprehension erlaubt es uns Listen durch eine andere Notation zu erstellen. Wir können uns es im Grunde als einzeiligen for Loop innerhalb von Klammer vorstellen. Ein einfaches Beispiel:

```python
# Jeden Buchstaben eines Strings erfassen
lst = [x for x in 'Wort']
```

```python
# Check
lst
```

Das ist das Grundkonzept von list comprehension. Wenn ihr euch etwas mit mathematischer Notation auskennt, sollte euch dieses Format bekannt vorkommen. Zum Beispiel: x^2; x in {0,1,2,...,10}

Lasst uns noch ein paar weitere Beispiel betrachten:

```python
# Quadradzahlen eines range() in eine Liste schreiben
lst = [x**2 for x in range(1,11)]
```

```python
# Check
lst
```

```python
# Check for even numbers in a range
lst = [x for x in range(11) if x % 2 == 0]
```

```python
# Check
lst
```

```python
# Convert Celsius to Fahrenheit
celsius = [0,10,20.1,34.5]

fahrenheit = [ ((float(9)/5)*temp + 32) for temp in celsius ]

# Check
fahrenheit
```

```python
lst = [ x**2 for x in [x**2 for x in range(11)]]
# Check
lst
```
